```python
import numpy as np 

possibilities = range(3,19)
occurences = np.array([1,3,6,10,15,21,25,27,27,25,21,15,10,6,3,1])
```

```python
import matplotlib.pyplot as plt

plt.rc('figure', figsize=(15, 9))


fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot(possibilities, occurences / sum(occurences))
ax1.set_title('Frequency distribution three dice')
ax1.set_xlabel('sum of dice')
ax1.set_ylabel('frequency')

ax2 = fig.add_subplot(2,2,2)
ax2.plot(possibilities, (occurences / sum(occurences)).cumsum())
ax2.set_title('Cumulative Frequency distribution three dice')
ax2.set_xlabel('sum of dice')
ax2.set_ylabel('cumulative frequency')
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

mu = 200
sigma = 25
n_bins = 50
x = np.random.normal(mu, sigma, size=100)

fig, ax = plt.subplots(figsize=(8, 4))

# plot the cumulative histogram
n, bins, patches = ax.hist(x, n_bins, density=True, histtype='step',
                           cumulative=True, label='Empirical')

# Add a line showing the expected distribution.
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
y = y.cumsum()
y /= y[-1]

ax.plot(bins, y, 'k--', linewidth=1.5, label='Theoretical')

# Overlay a reversed cumulative histogram.
ax.hist(x, bins=bins, density=True, histtype='step', cumulative=-1,
        label='Reversed emp.')

# tidy up the figure
ax.grid(True)
ax.legend(loc='right')
ax.set_title('Cumulative step histograms')
ax.set_xlabel('Annual rainfall (mm)')
ax.set_ylabel('Likelihood of occurrence')

plt.show()
```

```python
# Aggregate the counts
# or df.apply(sum, axis=0)

head_widths = df.sum(axis=0)
head_lengths = df.sum(axis=1)
```

```python
# Aggregator functions

def calc_mean(s):
    total = 0
    for i in s.index:
        total += i * s[i]    # label is average weight, value is count
    return total / s.sum()

def calc_median(s):
    
    # Get column label for cumulative position n
    def get_pos(n):
        for idx, pos in enumerate(s.cumsum()):
            if pos >= n:
                return s.index[idx]                                     
  
    int_sum = int(s.sum())
    if int_sum % 2 == 0:
        return (get_pos(int_sum / 2 - 1) + get_pos(int_sum / 2)) / 2
    else:
        return get_pos(int_sum - 1) / 2    
    
```

```python
# 4.3 calculate the formula for rat sample variance by hand and with Pandas/.
```

```python
rat_litters = pd.Series([7,33,58,116,125,126,121,107,56,37,25,4], index=range(1,13))
```

```python
rat_litters
```

```python
def sample_mean(s):
    return s.sum() / len(s)

def sample_variance(s):
    sum_of_squares = (s ** 2).sum()
    correction = len(s) * (sample_mean(s) **2)
    # correction = (s.sum()**2) / len(s)         # alternative computation
    return (sum_of_squares - correction) / len(s)   
    
```

```python
sample_mean(rat_litters)
```

```python
sample_variance(rat_litters)
```
